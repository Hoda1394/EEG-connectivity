# Reading data file

```python
file_loc = 'Data' # The EEG text file should be in Data folder
file_name = 'PAT36.txt' # Name of the EEG text file
patient_status = 'Abnormal' # Status of the patient
eeg_placement = 'BaptistLocation10-20.txt' # 2 options are available Baptist and Miami Children Hospital
sfreq = 200 # Sampling frequency
PATH = file_loc+'/'+file_name
pd.read_table(PATH,header = None).iloc[0]

col_remove = [0,1,5,11,18,24] # columns to be removed
patient = c_r.text_to_raw(PATH,eeg_placement,sfreq,col_remove)
```

# Plotting data

```python
%matplotlib
mne.viz.plot_raw(patient,scalings=dict(eeg=60e-3),n_channels=19,show = False,title=file_name,highpass=0.5, lowpass=50)
```

# Segment selection

```python
# Select windown with t_min and t_max
# Can be multiple segments as shown
time_stamp_segments =[
    [0,7],
    [23,40]
]
```

# Saving the segment to raw fif file

```python
patient_info = []
for i in range(len(time_stamp_segments)):
    temp_patient = patient.copy()
    temp_patient.crop(tmin = time_stamp_segments[i][0],tmax = time_stamp_segments[i][1])
    rawFifFileName = file_name.replace('.txt','_')+str(i)+'_raw.fif'
    
    #If raw_fif folder is not existed, create one
    if not os.path.exists('raw_fif'):
        os.makedirs('raw_fif')
    
    #save .fif file into raw_fif folder
    temp_patient.save('raw_fif/'+rawFifFileName,overwrite=True)
    patient_info.append([i,rawFifFileName,time_stamp_segments[i][0],time_stamp_segments[i][1],patient_status])
    
```

```python
columns_name = ['Segment','Name','t_min','t_max','Patient_Status']
patient_info_df = pd.DataFrame(patient_info,columns=columns_name)
patient_info_df
```

# Export table to csv

```python
if not os.path.exists('Info'):
    os.makedirs('Info')
patient_info_df.to_csv('Info/'+rawFifFileName.replace('.fif','')+'.csv')
```

# Calculating the connectivity

```python
file_loc = 'raw_fif/'
file_name = 'PAT36_0_raw.fif'
PATH = file_loc+file_name
print "       "+PATH

# Read raw.fif to mne file
patient = mne.io.read_raw_fif(PATH)
```

```python
def z_score_transform(raw_data):
    mu = np.mean(raw_data, axis=1)
    std = np.std(raw_data,axis=1)
    data = (raw_data - mu[:, None])/std[:,None]
    return data

raw_data = patient.get_data()[:,:]
# Normalized raw_data from epoch by z score transformation
data = z_score_transform(raw_data)
method_mvar = 'vm'
# assign static class cp.Mvar to variable mv
mv = cp.Mvar
# find best model order using Vieira-Morf algorithm
best, crit = mv.order_akaike(data, 15, method_mvar)
# Use connectivipy dataframe
data_df = cp.Data(data, fs=patient.info['sfreq'], chan_names=patient.ch_names)
# Fit model
data_df.fit_mvar(best, method_mvar)
# Compute ar vr coefficient
ar, vr = data_df.mvar_coefficients

# Compute PDC
#pdc_values = data_df.conn('dtf')
#pdc_significance = data_df.significance(Nrep=100, alpha=0.05)

# Compute gDTF
gdtf_values = data_df.conn('gdtf')
gdtf_significance = data_df.significance(Nrep=100, alpha=0.05)

```
